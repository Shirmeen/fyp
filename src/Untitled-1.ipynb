# Import Required Libraries
Import the necessary libraries, such as sqlite3 or SQLAlchemy.

```python
# Import the necessary libraries for database connection
import sqlite3
from sqlalchemy import create_engine
```

# Set Up Database Connection
Establish a connection to the database using the appropriate connection string.

```python
# Establish a connection to the database using the appropriate connection string
# For SQLite, we will connect to a local database file named 'example.db'

# Using sqlite3
conn = sqlite3.connect('example.db')
cursor = conn.cursor()

# Using SQLAlchemy
engine = create_engine('sqlite:///example.db')
connection = engine.connect()

# Verify the connection by executing a simple query
cursor.execute("SELECT sqlite_version();")
print("SQLite version:", cursor.fetchone())

# Close the connections
cursor.close()
conn.close()
connection.close()
```

# Execute SQL Queries
Write and execute SQL queries to interact with the database.

```python
# Re-establish the connection to the database
conn = sqlite3.connect('example.db')
cursor = conn.cursor()

# Function to execute a SQL query and fetch results
def execute_query(query):
    cursor.execute(query)
    return cursor.fetchall()

# Example: Create a table
create_table_query = """
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    email TEXT UNIQUE
);
"""
execute_query(create_table_query)

# Example: Insert data into the table
insert_data_query = """
INSERT INTO users (name, age, email) VALUES
('Alice', 30, 'alice@example.com'),
('Bob', 25, 'bob@example.com'),
('Charlie', 35, 'charlie@example.com');
"""
execute_query(insert_data_query)
conn.commit()  # Commit the transaction

# Example: Select data from the table
select_query = "SELECT * FROM users;"
results = execute_query(select_query)

# Print the results
for row in results:
    print(row)

# Close the connection
cursor.close()
conn.close()
```

# Fetch and Display Results
Fetch the results of the queries and display them using pandas or other visualization libraries.

```python
# Import pandas for data manipulation and visualization
import pandas as pd

# Fetch the results of the queries and display them using pandas
def fetch_and_display_results(query):
    results = execute_query(query)
    df = pd.DataFrame(results, columns=['id', 'name', 'age', 'email'])
    return df

# Example: Fetch and display the results of the 'SELECT' query
select_query = "SELECT * FROM users;"
df_results = fetch_and_display_results(select_query)
print(df_results)

# Close the connection
cursor.close()
conn.close()
```
